```python
import requests
import json
from bs4 import BeautifulSoup
import re
import pandas as pd
import time

headers={

'User-Agent': "Referer": "}

```

```python
companies=[]
jobnames=[]
jobdescs=[]
joblinks=[]
acceptRoles=[]
workExps=[]
edus=[]
otherreqs=[]
majors=[]
welfares=[]
industries=[]
specialties=[]
specialtyall=[]

keyword=input('請輸入關鍵字:')
#爬取多頁
url='https://www.104.com.tw/jobs/search/?ro=0&kwop=7&keyword='+keyword+'&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&order=12&asc=0&page='+'0'+'&mode=s&jobsource=2018indexpoc'
res=requests.get(url=url,headers=headers)
soup=BeautifulSoup(res.text,'html.parser')
totalp=int(str(soup.select('script')[13]).split()[-7].split(',')[-2].split(':')[-1])
print(totalp)

for p in range(1,totalp):
    print(p)
    time.sleep(5)
    url='https://www.104.com.tw/jobs/search/?ro=0&kwop=7&keyword='+keyword+'&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&order=12&asc=0&page='+str(p)+'&mode=s&jobsource=2018indexpoc'
    res=requests.get(url=url,headers=headers)
    soup=BeautifulSoup(res.text,'html.parser')
    jobsinapage=len(soup.select('article.b-block--top-bord.job-list-item.b-clearfix.js-job-item'))
    print(jobsinapage)
    if p==1:
        for i in range(10,29):
            jobList=soup.select('article.b-block--top-bord.job-list-item.b-clearfix.js-job-item')[i].text.replace(' ','').split('\n')
            company=list(filter(None,jobList))[2]
            if company in ['104外包網','【企業委託代徵】104獵才顧問中心','104家教網']:
                pass
            else:
                #company=list(filter(None,jobList))[2]
                jobname=list(filter(None,jobList))[1]
                joblink=soup.select('article.b-block--top-bord.job-list-item.b-clearfix.js-job-item')[i].find('a').get('href')[2:]
                print(company)
                print(jobname)
                print('https://'+joblink)


                url='https://www.104.com.tw/job/ajax/content/'+joblink.split('/')[-1]
                res=requests.get(url=url,headers=headers)
                jsonDict=json.loads(res.text)
                #acceptRole=jsonDict['data']['condition']['acceptRole']['role'][0]['description']
                workExp=jsonDict['data']['condition']['workExp']
                edu=jsonDict['data']['condition']['edu']
                otherreq=jsonDict['data']['condition']['other']
                major=jsonDict['data']['condition']['major']
                welfare=jsonDict['data']['welfare']['welfare']
                industry=jsonDict['data']['industry']

                specialtyList=jsonDict['data']['condition']['specialty']
                specialty=[]

                for i in specialtyList:
                    specialty.append(i['description'])
                    specialtyall.append(i['description']) 


                companies.append(company)
                jobnames.append(jobname)
                jobdescs.append(jobdesc)
                joblinks.append(joblink)
                acceptRoles.append(acceptRole)
                workExps.append(workExp)
                edus.append(edu)
                otherreqs.append(otherreq)
                majors.append(major)
                welfares.append(welfare)
                industries.append(industry)
                specialties.append(specialty)              
                print('====')
    else:
        for i in range(0,20):
            jobList=soup.select('article.b-block--top-bord.job-list-item.b-clearfix.js-job-item')[i].text.replace(' ','').split('\n')
            company=list(filter(None,jobList))[2]
            if company in ['104外包網','【企業委託代徵】104獵才顧問中心','104家教網']:
                pass
            else:
                jobname=list(filter(None,jobList))[1]
                joblink=soup.select('article.b-block--top-bord.job-list-item.b-clearfix.js-job-item')[i].find('a').get('href')[2:]
                print(company)
                print(jobname)
                print('https://'+joblink)


                url='https://www.104.com.tw/job/ajax/content/'+joblink.split('/')[-1]
                res=requests.get(url=url,headers=headers)
                jsonDict=json.loads(res.text)
                #acceptRole=jsonDict['data']['condition']['acceptRole']['role'][0]['description']
                workExp=jsonDict['data']['condition']['workExp']
                edu=jsonDict['data']['condition']['edu']
                otherreq=jsonDict['data']['condition']['other']
                major=jsonDict['data']['condition']['major']
                welfare=jsonDict['data']['welfare']['welfare']
                industry=jsonDict['data']['industry']

                specialtyList=jsonDict['data']['condition']['specialty']
                specialty=[]

                for i in specialtyList:
                    specialty.append(i['description'])
                    specialtyall.append(i['description']) 


                companies.append(company)
                jobnames.append(jobname)
                jobdescs.append(jobdesc)
                joblinks.append(joblink)
                acceptRoles.append(acceptRole)
                workExps.append(workExp)
                edus.append(edu)
                otherreqs.append(otherreq)
                majors.append(major)
                welfares.append(welfare)
                industries.append(industry)
                specialties.append(specialty)              
                print('====')



        
        
df=pd.DataFrame(columns=['公司','職缺','工作內容','工作經歷','學歷要求','科系要求','其他條件','工作技能','福利','產業','連結','Python'])
df['公司']=companies
df['職缺']=jobnames
df['工作內容']=jobdescs
#df['條件身分']=acceptRoles
df['工作經歷']=workExps
df['學歷要求']=edus
df['科系要求']=majors
df['其他條件']=otherreqs
df['工作技能']=specialties
df['福利']=welfares
df['產業']=industries
df['連結']=joblinks
df['Python']=df['工作技能'].apply(lambda x: 1 if 'Python' in x else 0)
df['R']=df['工作技能'].apply(lambda x: 1 if 'R' in x else 0)
df['MS SQL']=df['工作技能'].apply(lambda x: 1 if 'MS SQL' in x else 0)
df['PL/SQL']=df['工作技能'].apply(lambda x: 1 if 'PL/SQL' in x else 0)
df['hadoop']=df['工作技能'].apply(lambda x: 1 if 'hadoop' in x else 0)
df['Excel']=df['工作技能'].apply(lambda x: 1 if 'Excel' in x else 0)
df['PowerPoint']=df['工作技能'].apply(lambda x: 1 if 'PowerPoint' in x else 0)
df['Java']=df['工作技能'].apply(lambda x: 1 if 'Java' in x else 0)
df['C++']=df['工作技能'].apply(lambda x: 1 if 'C++' in x else 0)
df['C#']=df['工作技能'].apply(lambda x: 1 if 'C#' in x else 0)
df['C']=df['工作技能'].apply(lambda x: 1 if 'C' in x else 0)
df['Linux']=df['工作技能'].apply(lambda x: 1 if 'Linux' in x else 0)
df['Google Analytics']=df['工作技能'].apply(lambda x: 1 if 'Google Analytics' in x else 0)

```
